### Algorithm: 1



**1. Set up the following indicators on the 5-minute EURUSD chart:**

    5-period Exponential Moving Average (EMA5)
    20-period Simple Moving Average (SMA20)
    Stochastic Oscillator (14, 3, 3)
    Relative Strength Index (RSI, 14-period)
    Bollinger Bands (20-period, 2 standard deviations)
    Average True Range (ATR, 14-period)
    
**2. Buy conditions:**

    EMA5 crosses above SMA20.
    Stochastic Oscillator %K line crosses above the %D line and is below 30 (oversold level).
    RSI is above 50 and rising.
    The price is near or below the lower Bollinger Band.
    
**3. Sell conditions:**

    EMA5 crosses below SMA20.
    Stochastic Oscillator %K line crosses below the %D line and is above 70 (overbought level).
    RSI is below 50 and falling.
    The price is near or above the upper Bollinger Band.
    
**4. Stop-loss and take-profit:**

    Calculate the ATR value at the time of trade entry.
    Set the stop-loss at 1.5 x ATR below the entry price for a long position, or 1.5 x ATR above the entry price for a short position.
    Set the take-profit at 1 x ATR above the entry price for a long position, or 1 x ATR below the entry price for a short position.

#### Buy conditions:

    EMA5 crosses above SMA20.
    Stochastic Oscillator %K line crosses above the %D line and is below 30 (oversold level).
    RSI is above 50 and has recently crossed above its 30-period EMA.
    The price is near or below the lower Bollinger Band and has recently closed above it.
    Confirm the trade with higher timeframe trend alignment (e.g., 15-minute or 1-hour chart) to increase the probability of success.
    
#### Sell conditions:

    EMA5 crosses below SMA20.
    Stochastic Oscillator %K line crosses below the %D line and is above 70 (overbought level).
    RSI is below 50 and has recently crossed below its 30-period EMA.
    The price is near or above the upper Bollinger Band and has recently closed below it.
    Confirm the trade with higher timeframe trend alignment (e.g., 15-minute or 1-hour chart) to increase the probability of success.
    
#### Stop-loss and take-profit:

    Calculate the ATR value at the time of trade entry.
    Set the stop-loss at 1.5 x ATR below the entry price for a long position, or 1.5 x ATR above the entry price for a short position.
    Set the take-profit at 2 x ATR above the entry price for a long position, or 2 x ATR below the entry price for a short position.
    Optionally, you can use a trailing stop based on a multiple of ATR (e.g., 1 x ATR) to lock in profits as the trade moves in your favor.

#### Sell conditions:

    EMA9 crosses below EMA21.
    CCI crosses below 100 from above, indicating a shift from an overbought condition to a more bearish market sentiment.
    ADX is above 25, suggesting a strong trend.

#### Stop-loss and take-profit:

    Calculate the ATR value at the time of trade entry.
    Set the stop-loss at 1.5 x ATR below the entry price for a long position, or 1.5 x ATR above the entry price for a short position.
    Set the take-profit at 2 x ATR above the entry price for a long position, or 2 x ATR below the entry price for a short position.
