```python


# Add buy_sl and buy_tp columns to the DataFrame
df['buy_sl'] = 99
df['buy_tp'] = 99

def get_buy_sl_tp(openprice, stoploss=300, takeprofit=600):
    """
    Calculate the stop loss and take profit values for a given open price.
    """
    sl = round((openprice - stoploss / 100000), 5)
    tp = round((openprice + takeprofit / 100000), 5)
    return sl, tp

for index, row in df.iterrows():
    openprice = row['open']
    sl, tp = get_buy_sl_tp(openprice, stoploss=300, takeprofit=600)

    for subindex, subrow in df.loc[index:].iterrows():
        lowprice = subrow['low']
        highprice = subrow['high']

        if (sl > lowprice and tp > highprice):
            df.loc[index, 'buy_sl'] = 1
            df.loc[index, "buy_tp"] = 0
            break

        elif (lowprice > sl and highprice > tp):
            df.loc[index, 'buy_sl'] = 0
            df.loc[index, 'buy_tp'] = 1
            break

        # If both stop loss and take profit are hit on the same candle
        elif (lowprice < sl and highprice > tp):
            df.loc[index, 'buy_sl'] = 2
            df.loc[index, 'buy_tp'] = 2
            break
```
